```python
df['Latitude'] = df['Latitude'].replace(regex=',', value='.')
df["Latitude"] = df["Latitude"].astype('float')
df['Longitud'] = df['Longitud'].replace(regex=',', value='.')
df["Longitud"] = df["Longitud"].astype('float')
```

```python
geolocalizacoes = df[['Latitude', 'Longitud']]
```

```python
coordenadas = []
lat = geolocalizacoes['Latitude'].values

long = geolocalizacoes['Longitud'].values

mapa = folium.Map(location=[-23.1791, -45.8872],tiles='OpenStreetMap',zoom_start=12)
 
for la,lo in zip(lat,long):
    coordenadas.append([la,lo])
    
mapa.add_child(plugins.HeatMap(coordenadas))
```
